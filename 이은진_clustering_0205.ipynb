## 레이터 차트

```python
pd.DataFrame(df2.groupby(['cluster'])['p01_05', 'p06_10', 'p11_14', 'p15_17', 'p18_20', 'p21_24'].mean()).reset_index()
```

```python
from math import pi
from matplotlib.path import Path
from matplotlib.spines import Spine
from matplotlib.transforms import Affine2D

df = pd.DataFrame(df2.groupby(['cluster'])['p01_05', 'p06_10', 'p11_14', 'p15_17', 'p18_20', 'p21_24'].mean()).reset_index()
labels = df.columns[1:]
num_labels = len(labels)
    
angles = [x/float(num_labels)*(2*pi) for x in range(num_labels)] ## 각 등분점
angles += angles[:1] ## 시작점으로 다시 돌아와야하므로 시작점 추가
    
my_palette = plt.cm.get_cmap("Set2", len(df.index))
 
fig = plt.figure(figsize=(15,15))
fig.set_facecolor('white')
 
for i, row in df.iterrows():
    color = my_palette(i)
    data = df.iloc[i].drop('cluster').tolist()
    data += data[:1]
    
    ax = plt.subplot(3,3, i+1, polar=True)
    ax.set_theta_offset(pi / 2) ## 시작점
    ax.set_theta_direction(-1) ## 그려지는 방향 시계방향
    
    #ax.set_xticklabels(angles[:-1], labels, fontsize=13)
    plt.xticks(angles[:-1], labels, fontsize=13) ## x축 눈금 라벨
    ax.tick_params(axis='x', which='major', pad=15) ## x축과 눈금 사이에 여백을 준다.
 
    ax.set_rlabel_position(0) ## y축 각도 설정(degree 단위)
    plt.yticks([0,10,20,30,40,50],['0','10','20','30','40','50'], fontsize=10) ## y축 눈금 설정
    plt.ylim(0,55)
    
    ax.plot(angles, data, color=color, linewidth=2, linestyle='solid') ## 레이더 차트 출력
    ax.fill(angles, data, color=color, alpha=0.4) ## 도형 안쪽에 색을 채워준다.
    
    plt.title(row.cluster, size=20, color=color,x=-0.2, y=1.2, ha='left') ## 타이틀은 캐릭터 클래스로 한다.
 

plt.tight_layout(pad=5) ## subplot간 패딩 조절
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

df = pd.DataFrame(df2.groupby(['cluster'])['n_card',  'count/n_card','regular_percent',
       'weekdays_amount_avg', 'weekend_amount_avg', 'total_amount/n_days', 'holiday_amount/n_holiday'].mean()).reset_index()
labels = df.columns[1:]
num_labels = len(labels)


scaler = MinMaxScaler()
output = df.drop(['cluster'],axis=1)
scaler.fit(output)
output = scaler.transform(output)

output = pd.DataFrame(output, columns=labels, index=list(df.index.values))
output['cluster'] = df['cluster']
    
angles = [x/float(num_labels)*(2*pi) for x in range(num_labels)] ## 각 등분점
angles += angles[:1] ## 시작점으로 다시 돌아와야하므로 시작점 추가
    
my_palette = plt.cm.get_cmap("Set2", len(output.index))
 
fig = plt.figure(figsize=(20,20))
fig.set_facecolor('white')
 
for i, row in output.iterrows():
    color = my_palette(i)
    data = output.iloc[i].drop('cluster').tolist()
    data += data[:1]
    
    ax = plt.subplot(3,3, i+1, polar=True)
    ax.set_theta_offset(pi / 2) ## 시작점
    ax.set_theta_direction(-1) ## 그려지는 방향 시계방향
    
    #ax.set_xticklabels(angles[:-1], labels, fontsize=13)
    plt.xticks(angles[:-1], labels, fontsize=13) ## x축 눈금 라벨
    ax.tick_params(axis='x', which='major', pad=15) ## x축과 눈금 사이에 여백을 준다.
 
    ax.set_rlabel_position(0) ## y축 각도 설정(degree 단위)
    plt.yticks([0,0.2,0.4,0.6,0.8],['0','1','2','3','4'], fontsize=10) ## y축 눈금 설정
    plt.ylim(0,1)
    
    ax.plot(angles, data, color=color, linewidth=2, linestyle='solid') ## 레이더 차트 출력
    ax.fill(angles, data, color=color, alpha=0.4) ## 도형 안쪽에 색을 채워준다.
    
    plt.title(row.cluster, size=20, color=color,x=-0.2, y=1.2, ha='left') ## 타이틀은 캐릭터 클래스로 한다.
 

plt.tight_layout(pad=5) ## subplot간 패딩 조절
plt.show()
```
